```python
import torch
from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer, pipeline

# Path to saved peft adapter model
# peft_model_id = args.output_dir # or
peft_model_id = "DrishtiSharma/doplhin-2.1-mistral-7b-orpo-ultrafeedback-binarized-preferences"

# Load Model with PEFT adapter
model = AutoPeftModelForCausalLM.from_pretrained(
  peft_model_id,
  device_map="cuda",
  torch_dtype=torch.float16
)
tokenizer = AutoTokenizer.from_pretrained(peft_model_id)
# load into pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompts = [
    "Calculate the area of a circular pond if the diameter is 10 feet. How many square feet of liner material will you need to cover the bottom of the pond?",
    "A mysterious ancient manuscript is found in your attic, describing a potion made from three ingredients: dragon's blood, phoenix feathers, and a rare herb found only in the moonlight. What could this potion do, according to legend?",
    "If you were to design a robot to help with household chores, what three unique features would it have to make your life easier?",
    "You discover a new planet in the solar system with a unique property: its gravity is twice that of Earth's. What would life be like on this planet, and how would it affect human physiology?",
    "In a world where everyone has a superpower based on their personality, what would yours be and how would you use it to benefit society?",
    "If you could create a new holiday that everyone in the world would celebrate, what would it be called and how would people celebrate it?",
    "You find a map leading to a hidden treasure buried by pirates centuries ago. What do you find when you follow the map, and how does it change your life?",
    "A secret society has been protecting a powerful artifact for millennia. You've just been initiated as a member. What is the artifact, and why is it so important to keep it safe?",
    "Design a sustainable city of the future. What innovative technologies and systems will you incorporate to ensure its residents live in harmony with the environment?",
    "You've been given the power to communicate with animals for a day. Which animals would you talk to first, and what would you ask them?"
]
```

```python
for prompt in prompts:
  messages = pipe.tokenizer.apply_chat_template([{"role":"user", "content": prompt}], tokenize=False)
  outputs = pipe(prompt, max_new_tokens=2048, do_sample=True, temperature=1.0, top_k=50, top_p=0.9, eos_token_id=tokenizer.eos_token_id, pad_token_id=tokenizer.pad_token_id)
  print(f"**Prompt**:\n{prompt}\n")
  print(f"**Generated Answer**:\n{outputs[0]['generated_text'][len(prompt):].strip()}")
  print("===" * 10)
```
